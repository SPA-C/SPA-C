```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
from matplotlib.colors import ListedColormap
from matplotlib.patches import ConnectionPatch
import matplotlib.patches as mpatches
import matplotlib.colors as plc
from matplotlib.colors import ListedColormap
import cooler
import string
from numpy.lib.stride_tricks import as_strided
```

```python
fig, ax = plt.subplots(1, 1, figsize=(4,3), dpi=600)

fs = 7
plt.rcParams.update({'font.size': fs, 'legend.fontsize':fs, 'figure.labelsize':fs, 'xtick.labelsize': fs, 'ytick.labelsize': fs})

# building the array
arr = np.zeros((20, 20))
arr[:10,:10] = 1 # Region 1
arr[10:, 10:] = 2 # Region 2
arr[:10,10:] = 3 # Region 1|2
arr[10:,:10] = 3 # Region 2|1

# Custom CMAP
colors = {"Region A": "#277DA1", "Region B": "#F9C74F", "Adjacent region": "#90BE6D"}
cmap = ListedColormap(
    list(colors.values())
)

ax.matshow(arr, cmap=cmap)
ax.tick_params(
            axis='both',          
            which='both',         
            bottom=False,         
            top=False,            
            left=False,           
            right=False,          
            labelbottom=False,    
            labeltop=False,       
            labelleft=False,      
            labelright=False      
        )

rectangles, labels = [], []
for key, value in colors.items():
    rectangles.append(mpatches.Patch(
        color=value,
        alpha=1,
        fill=True,
        label=key,
        linewidth=1
    ))
    labels.append(key)

ax.text(-.8, -.5, "0", fontsize=fs, ha="right", va="bottom")
for pos in [10, 20]:
    ax.text(-.8, pos-.5, pos, fontsize=fs, ha="right", va="center")
    ax.text(pos-.5, -1, pos, fontsize=fs, ha="center", va="center")

fig.legend(rectangles, labels, loc='right', ncol=1, bbox_to_anchor=(0.97, 0.5), frameon=False, facecolor='white', edgecolor='none')
plt.tight_layout(rect=(0,0.05,0.7,1))

plt.savefig(f"SUPP.PAPER.ImputStruct.png")
plt.show()
```
