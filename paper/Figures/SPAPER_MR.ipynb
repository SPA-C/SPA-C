# MR vs no MR figure/table

```python
import pandas as pd

file = "../Results/SPAPER_MR.csv"

df = pd.read_csv(file, sep="\t", header=[0, 1], index_col=0)
df
```

```python
import numpy as np
datasets = [tup[0] for tup in df.columns.tolist()[::2]]
datasets

sdf = df.loc[["NGA50", "NGA90", "auNGA", "# misassemblies"]]
for column in sdf.columns:
    sdf[column] = sdf[column].astype(float)/1000000 # Convert to Mb
sdf.loc["# misassemblies"] = (sdf.loc["# misassemblies"]*1000000).astype(int)  
sdf
```

```python
for ds in datasets:
    sdf[(ds, "Difference")] = (( (sdf[(ds, "SPA-C.MR10")] - sdf[(ds, "SPA-C.5k")]) / sdf[(ds, "SPA-C.5k")] ) * 100).astype(str).apply(lambda val: f"{float(val):.2f}%")
    
metrics = ["SPA-C.MR10", "SPA-C.5k", "Difference"]
new_columns = pd.MultiIndex.from_product([datasets, metrics], names=["Dataset", None])
sdf = sdf.reindex(columns=new_columns).T.rename(index={"SPA-C.MR10": "SPA-C", "SPA-C.5k":"SPA-C@5k"})

italic = lambda x: f"\\textit{{{x}}}"
sdf.rename(
    index={
        "AG.DtoL": italic("A. gentilis"), 
        "AT.DtoL": italic("A. thaliana"), 
        "BP.DtoL": italic("B. physalus"), 
        "Dro.DtoL": italic("D. histrio"), 
        "HG002.B10M": "HG002", 
        "Vul.DtoL": italic("V. vulpes")},
    columns={"# misassemblies": "\# misassemblies", "NGA50": r"NGA50$^{1}$", r"NGA90": r"NGA90$^{1}$", r"auNGA":r"auNGA$^{1}$"},
    inplace=True
)

sdf
```
