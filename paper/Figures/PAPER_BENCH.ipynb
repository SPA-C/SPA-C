# Librairies and file loading

```python
import pandas as pd
import numpy as np
```

## Quast metrics

```python
file = "../Results/PAPER_BENCH.csv"
```

```python
# Parsing tables within the CSV sheet
full = pd.read_csv(file, sep="\t", nrows=54, header=[0, 1], index_col=0, decimal=',')
ranks = pd.read_csv(file, sep="\t", skiprows=58, nrows=7, header=[0, 1], index_col=0, decimal=',')

main = pd.concat([full, ranks.loc[["Average Rank"]]])
main = main.swaplevel(0, 1, axis=1).sort_index(axis=1).replace({np.nan: -1})

main
```

```python
# Creating table for the paper
table = main.loc[["# misassemblies", "NGA50", "NGA90", "auNGA", "Average Rank"], ["YaHS", "Pin_hic", "3DDNA.FINAL", "AutoHic", "SPA-C.MR10"]].T
italic = lambda x: f"\\textit{{{x}}}"
table.rename(
    columns = {"Average Rank": r"Dataset rank$^{2}$", "# misassemblies": "\# misassemblies", "NGA50": r"NGA50$^{1}$", r"NGA90": r"NGA90$^{1}$", r"auNGA":r"auNGA$^{1}$"}, 
    index = {"Pin_hic": r"Pin\_hic","3DDNA.FINAL": "3D-DNA", "SPA-C.MR10": "SPA-C", "AG.DToL": italic("A. gentilis"), "AT.DtoL": italic("A. thaliana"), "BP.DtoL": italic("B. physalus"), "Dro.DtoL": italic("D. histrio"), "HG002.B10M": "HG002", "Vul.DToL": italic("V. vulpes")},
    inplace = True
)
for col in [r"NGA50$^{1}$", r"NGA90$^{1}$", r"auNGA$^{1}$"]:
    table[col] = table[col].astype(int) / 1000000
table["\# misassemblies"] = table["\# misassemblies"].astype(int)
table.index.names = ['Tool', 'Dataset']

## Computing global rank across datasets
avg_rank = table.groupby("Tool")[[r"Dataset rank$^{2}$"]].mean().rename(columns = {r"Dataset rank$^{2}$": "Global rank$^{3}$"})
table = table.reset_index(drop=False).merge(avg_rank, on='Tool').set_index(['Tool', 'Dataset']).apply(lambda x: x.apply(lambda y: np.nan if y < 0 else y) if np.issubdtype(x.dtype, np.number) else x)

table
```

## Time metrics

```python
tfile = "../Results/PAPER_TIME.csv"
time_df = pd.read_csv(tfile, sep="\t", header=0, index_col=None).set_index(['Step', 'Dataset'])
time_df
```

```python
import numpy as np

def time_to_seconds(time_str):
    if '-' in time_str:
        days, time_part = time_str.split('-', 1)
        days = int(days)
    else:
        days = 0
        time_part = time_str

    hh, mm, ss = map(int, time_part.split(':'))

    total_seconds = days * 86400 + hh * 3600 + mm * 60 + ss
    return total_seconds

def compute_all_time(row, df):
    row.dropna(inplace=True, axis=1)
    cols = row.columns
    steps = np.unique([col.split(".")[0] for col in cols])

    total_time = 0
    for step in steps:
        step_time = time_to_seconds(row.reset_index()[f"{step}.time"][0]) * int(row.reset_index()[f"{step}.threads"][0])
        total_time += step_time
        #print(total_time, step_time)

    if row.index[0][0] == "AutoHic":
        step_time = time_to_seconds(df.loc[("3D-DNA", row.index[0][1]), "Tool.time"])*df.loc[("3D-DNA", row.index[0][1]), "Tool.threads"]
        total_time += step_time
        print(total_time, step_time)
    
    return total_time
```

```python
time_df["CPU time (s)"] = ([int(compute_all_time(time_df.iloc[[k]], time_df)) for k in range(len(time_df))])
time_df["CPU time (h)"] = (time_df["CPU time (s)"]//3600).astype(int)

time_df = time_df.reset_index().rename(
    columns = {"Step": "Tool"}, 
).set_index(['Tool', 'Dataset']).rename(
    index = {"Pin_hic": r"Pin\_hic", "AG.DToL": italic("A. gentilis"), "AT.DtoL": italic("A. thaliana"), "BP.DtoL": italic("B. physalus"), "Dro.DtoL": italic("D. histrio"), "HG002.B10M": "HG002", "Vul.DToL": italic("V. vulpes")}
)

time_df
```

```python
PTab = pd.concat([table, time_df[["CPU time (h)"]]], axis=1).iloc[:,[0, 1, 2, 3, 6, 4, 5]]
PTab
```
