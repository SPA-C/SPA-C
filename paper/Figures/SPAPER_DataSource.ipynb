```python
import pandas as pd

file = "../Results/SPAPER_Data.csv"

df = pd.read_csv(file, sep="\t", header=0, index_col=None)
df = df[["Dataset", "Link", "Assembly download link", "Hi-C reads download links"]]
df
```

```python
Lcode = df.to_latex(float_format="%.2f")
Finalcode="".join([
    "\\begin{table*}[t]\n\\caption{Datasets and associated data sources.}\n\\tabcolsep=0pt%%\n",
    Lcode.replace(
        "tabular", "tabular*"
    ).replace(
        "{lllll}", r"{\textwidth}{@{\extracolsep{\fill}}lccc@{\extracolsep{\fill}}}"
    ).replace(
        "Link", "Main link"
    ).replace(
        "_", r"\_"
    ).replace(
        "%", r"\%"
    ),
    "\\label{stable:datasource}\n\\end{table*}"
])

print(Finalcode)
```
